```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
```

```python
df = pd.read_csv("bankruptcy-prevention.csv")
df
```

```python
df.columns
```

```python
# replace "bankruptcy" by '0' and "non-bankruptcy" as '1'

df["class"] = df["class"].replace("bankruptcy",0)
df["class"] = df["class"].replace("non-bankruptcy",1)
```

```python
df
```

```python
# Converting datset into array format

data = df.values
data
```

```python
# Spliting the data into dependent amd independent variable

x = df.iloc[: , :-1]
y = df.iloc[:,-1]
```

```python
# Train - Test split

from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x,y ,
                                   random_state = 42, 
                                   test_size = 0.3)
#here we split our whole dataset in 70% - 30% split. 
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression(penalty = 'l2' , max_iter = 100 , C = 1)
LR.fit(x_train , y_train)
pred_LR = LR.predict(x_test)
```
